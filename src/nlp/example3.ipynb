You can use a **dispersion_plot** to understand the frequency and pattern of usage of a word in a corpus.

```python
import nltk.book as book
```

```python
book.text4.dispersion_plot(["citizens", "democracy", "freedom", "duties", "America"])
```

```python
book.text4.dispersion_plot(["America", "Canada", "NAFTA", "Politics", "War"])
```

```python
book.text4.dispersion_plot(["liberty", "constitution"])
```

```python
book.text4.dispersion_plot(["citizens", "democracy", "freedom", "duties", "liberty", "constitution","America", "politics", "war"])
```
