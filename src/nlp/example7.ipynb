# Frequency Distribution in NLTK

### Let's find the 50 most frequently used words in "Moby Dick"

```python
import nltk.book as book
```

```python
frequency_distribution_moby_dick = book.FreqDist(book.text1)
print(frequency_distribution_moby_dick)
```

```python
# the 80 most common 
frequency_distribution_moby_dick.most_common(80)
```

```python
# Let's see how many times the word "boat" has been mentioned in Moby Dick
frequency_distribution_moby_dick["boat"]
```

```python
frequency_distribution_moby_dick.plot(50, cumulative=True)
```

### These 50 words together have appeared more than 120_000 times in this text (which is almost half of this book)

```python
# Let's see what words have been mentioned in Moby Dick only once
frequency_distribution_moby_dick.hapaxes()
```
