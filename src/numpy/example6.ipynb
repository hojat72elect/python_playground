Lets show up a normal distribution of data in matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

example_incomes = np.random.normal(100.0, 20.0, 10_000) # 10_000 random numbers centered around 100 with the standard deviation of 20
plt.hist(x=example_incomes, bins=50)
plt.show()
```

```python
# calculate the standard deviation
standard_deviation = example_incomes.std()
standard_deviation
```

```python
# variance is the square of the standard deviation
variance = example_incomes.var()
variance
```
