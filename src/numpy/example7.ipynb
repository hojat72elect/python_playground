In this notebook, we will have a look at different types of data distribution.

```python
# First do all the imports
import numpy
import matplotlib.pyplot
import scipy.stats
```

```python
"""
First : Uniform distribution
It just means there's a flat constant probability of a value occurring within a given range. 
"""
values = numpy.random.uniform(low=-10.0, high=10.0, size=100_000)
matplotlib.pyplot.hist(x=values, bins=50)
matplotlib.pyplot.show()
```

```python
"""
Second : Normal or Gaussian distribution
It is a bell shaped curve, with a mean and standard deviation.
"""
# first example of normal distribution
x = numpy.arange(start=-3, stop=3, step=0.001)
matplotlib.pyplot.plot(x, scipy.stats.norm.pdf(x))
```

```python
# Second example from a normal distribution
mu = 5.0 # The mean of the distribution
sigma = 2.0 # The standard deviation of the data

values = numpy.random.normal(mu, sigma, 10_000)
matplotlib.pyplot.hist(values, 50)
matplotlib.pyplot.show()
```

```python
"""
Third : Exponential distribution (or Power Law)
"""
x = numpy.arange(0, 10, 0.001)
matplotlib.pyplot.plot(x, scipy.stats.expon.pdf(x))
```
