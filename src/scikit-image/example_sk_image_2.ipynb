A simple example of how to apply any of the filters in "filters" module to a given picture.

```python
import skimage
import matplotlib.pyplot as plt

image = skimage.data.coins() # or any Numpy array!
edges = skimage.filters.sobel(image)

# Show Original and Edges side by side
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax[0].imshow(image, cmap='gray')
ax[0].axis('off')  # Hide the axis
ax[1].imshow(edges, cmap='gray')
ax[1].axis('off')  # Hide the axis


plt.imshow(edges, cmap='gray')
plt.axis('off')  # Hide the axis
plt.show()


```
